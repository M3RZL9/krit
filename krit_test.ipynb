```python
def algorithm():
    result = []
    for I in P_indicators:
        for R in P_resources:
            for Y in P_years:
                for i in T_values:
                    if i['Показатель'] == I and i['Энергетический ресурс'] == R and i['Год'] == Y:
                        i['Причина ошибки'] = 'Источник'
                        result.append(i)
                while True:
                    M_values = [i for i in result if i['Показатель'] == I
                                and i['Энергетический ресурс'] == R and i['Год'] == Y]
                    
                    units = [i['Единица измерения'] for i in M_values]
                    
                    M_calculated = [i for i in T_multiplication if i['Базовая ЕИ'] in units
                                    and i['Расчетная ЕИ'] not in units]
                    
                    M_based = [i for i in T_multiplication if i['Расчетная ЕИ'] in units
                               and i['Базовая ЕИ'] not in units]
                    
                    M_result = [i for i in T_convertation if i['Исходная ЕИ'] in units
                               and i['Результирующая ЕИ'] not in units]
                    
                    if len(M_calculated) + len(M_based) + len(M_result) == 0:
                        break
                    
                    if len(M_calculated) > 0:
                        for i in M_values:
                            for n in M_calculated:
                                if i['Единица измерения'] == n['Базовая ЕИ']:
                                    result.append({'Показатель' : I,
                                                   'Энергетический ресурс' : R, 
                                                   'Год' : Y, 
                                                   'Единица измерения' : n['Расчетная ЕИ'], 
                                                   'Значение': i['Значение'] * 10**n['Степень'],
                                                   'Причина ошибки': 'из %s в %s' % (n['Базовая ЕИ'], n['Расчетная ЕИ'])})
                    
                    if len(M_based) > 0:
                        for i in M_values:
                            for n in M_based:
                                if i['Единица измерения'] == n['Расчетная ЕИ']:
                                    result.append({'Показатель' : I,
                                                   'Энергетический ресурс' : R, 
                                                   'Год' : Y, 
                                                   'Единица измерения' : n['Базовая ЕИ'], 
                                                   'Значение': i['Значение'] * 10**(-n['Степень']),
                                                   'Причина ошибки': 'из %s в %s' % (n['Расчетная ЕИ'], n['Базовая ЕИ'])})

                    if len(M_result) > 0:
                            for i in M_values:
                                for n in M_result:
                                    if i['Единица измерения'] == n['Исходная ЕИ']:
                                        result.append({'Показатель' : I,
                                                    'Энергетический ресурс' : R, 
                                                    'Год' : Y, 
                                                    'Единица измерения' : n['Результирующая ЕИ'], 
                                                    'Значение': i['Значение'] * n['Коэффициент'],
                                                    'Причина ошибки': 'из %s в %s' % (n['Исходная ЕИ'], n['Результирующая ЕИ'])})
                                    
    return result


P_indicators = ['СКЭ']

P_resources = ['Природный газ']

P_years = [2017]

T_values = [
    { 'Показатель': 'СКЭ', 'Энергетический ресурс': 'Природный газ', 'Год': 2017, 'Единица измерения': 'Mtoe', 'Значение': 148.67 },
    { 'Показатель': 'СКЭ', 'Энергетический ресурс': 'Природный газ', 'Год': 2018, 'Единица измерения': 'Mtoe', 'Значение': 149.33 },
    { 'Показатель': 'СКЭ', 'Энергетический ресурс': 'Природный газ', 'Год': 2019, 'Единица измерения': 'Mtoe', 'Значение': 150.00 },
    { 'Показатель': 'СКЭ', 'Энергетический ресурс': 'Природный газ', 'Год': 2020, 'Единица измерения': 'Mtoe', 'Значение': 150.67 },
    { 'Показатель': 'СКЭ', 'Энергетический ресурс': 'Природный газ', 'Год': 2021, 'Единица измерения': 'Mtoe', 'Значение': 151.33 },]

T_multiplication = [
    { 'Расчетная ЕИ': 'Gft3ng', 'Степень': -9, 'Базовая ЕИ': 'ft3ng' },
    { 'Расчетная ЕИ': 'Gtce', 'Степень': -9, 'Базовая ЕИ': 'tce' },
    { 'Расчетная ЕИ': 'Mtoe', 'Степень': -6, 'Базовая ЕИ': 'toe' },
    { 'Расчетная ЕИ': 'MMbtu', 'Степень': -6, 'Базовая ЕИ': 'btu' },
    { 'Расчетная ЕИ': 'Twh', 'Степень': -12, 'Базовая ЕИ': 'wh' },
    { 'Расчетная ЕИ': 'Gwh', 'Степень': -9, 'Базовая ЕИ': 'wh' },
    { 'Расчетная ЕИ': 'Gj', 'Степень': -9, 'Базовая ЕИ': 'j' },
    { 'Расчетная ЕИ': 'Mboe', 'Степень': -6, 'Базовая ЕИ': 'boe' },
    { 'Расчетная ЕИ': 'Gm3ng', 'Степень': -9, 'Базовая ЕИ': 'm3ng' }]

T_convertation = [
    { 'Исходная ЕИ': 'tce', 'Коэффициент': 751.4768963, 'Результирующая ЕИ': 'm3ng' },
    { 'Исходная ЕИ': 'ft3ng', 'Коэффициент': 301.277062, 'Результирующая ЕИ': 'wh' },
    { 'Исходная ЕИ': 'MMbtu', 'Коэффициент': 1055.060005, 'Результирующая ЕИ': 'Mj' },
    { 'Исходная ЕИ': 'boe', 'Коэффициент': 5551365.225, 'Результирующая ЕИ': 'btu' },
    { 'Исходная ЕИ': 'toe', 'Коэффициент': 1.4285714, 'Результирующая ЕИ': 'tce' },
    { 'Исходная ЕИ': 'j', 'Коэффициент': 0.000277778, 'Результирующая ЕИ': 'wh' },
    { 'Исходная ЕИ': 'toe', 'Коэффициент': 7.1428, 'Результирующая ЕИ': 'boe' },
    { 'Исходная ЕИ': 'm3ng', 'Коэффициент': 35.958043, 'Результирующая ЕИ': 'ft3ng' },]

result = algorithm()
for i in result:
    print('%s у показателя "%s" в %d году составляет %f %s (%s)' %
          (i['Показатель'], i['Энергетический ресурс'], i['Год'], 
           i['Значение'], i['Единица измерения'], i['Причина ошибки']))
```
